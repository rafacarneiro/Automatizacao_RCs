```python
# Importando histórico de compras por material da ME2M

compras = pd.read_excel('compras - 2.xlsx')
compras
```

```python
# Importando contratos por material da ME3M

contratos = pd.read_excel('contratos - 2.xlsx')
contratos
```

```python
# Importando tabela criada com as Fórmulas de reajuste por material 

condicoes = pd.read_excel('condicoes - 2.xlsx')
condicoes
```

```python
# Importando tabela com os drivers de preço de cada material. Tabela com todos os índices já acumulados.

## O ideal é coletar os dados diretamente da internet, mas, foi coletado e tratado em excel apenas para agilidade na demonstração.

indices = pd.read_excel('indices - 2.xlsx')
indices
```

# <font color=orange>3) Aplicando as fórmulas de reajuste para atualização dos Preços</font>
***

```python
# Selecionando tabela de compras e selecionando o último pedido para cada código de material

df_compras = compras[{'Pedido', 'Material', 'Data Doc.', 'Fornecedor', 'Valor Unitario', 'UM', 'Frete'}]
ultimo_valor = df_compras.groupby(["Material"])['Pedido'].max().reset_index()
df_compras = pd.merge(ultimo_valor, df_compras, on=['Material'], how='left')


## Função para excluir pedidos anteriores ao último de cada material

def f(row):
    if row['Pedido_x'] == row['Pedido_y']:
        val = 0
    elif row['Pedido_x'] > row['Pedido_y']:
        val = 1
    else:
        val = -1
    return val
df_compras["retirar"] = df_compras.apply(f, axis=1)


### Ajustando tabela para melhor visualização dos dados

df_compras= df_compras.rename(columns={'Pedido_x': 'Pedido'})
df_compras = df_compras.query('retirar==0')
df_compras = df_compras.drop(columns=['Pedido_y', 'retirar'])
df_compras
```

```python
# Unificando os indicadores e a fórmula de reajuste dos materiais

df_preco_novo = df_compras[{'Material', 'Pedido', 'Data Doc.', 'Valor Unitario'}]
df_indices = indices[{'Data', 'Dolar F', 'IGPM F', 'Petro F', 'Soja F', 'Cobre F', 'Estanho F'}]
df_indices= df_indices.rename(columns={'Data': 'Data Doc.'})
df_preco_novo = pd.merge(df_preco_novo, df_indices, on=['Data Doc.'], how='left')
df_preco_novo = pd.merge(df_preco_novo, condicoes, on=['Material'], how='left')
df_preco_novo = df_preco_novo.drop(columns=['Grp Merc', 'Fórm. Reaj', 'Texto Breve', 'UM'])
df_preco_novo= df_preco_novo.rename(columns={'Fórm. Reaj.1': 'form_reaj'})
df_preco_novo
```

```python
# Aplicando a fórmula de reajuste com base na coluna "form_reaj" e adicionando coluna "preco_previsto"

fr001 = df_preco_novo.query('form_reaj=="FR-001"')
fr001["preco_previsto"] = ((fr001["Valor Unitario"] * 
fr001["Cobre F"])/100 + 
fr001["Valor Unitario"]).round(2)

fr002 = df_preco_novo.query('form_reaj=="FR-002"')
fr002["preco_previsto"] = ((fr002["Valor Unitario"] * 
fr002["Dolar F"])/100 + 
fr002["Valor Unitario"]).round(2)

fr003 = df_preco_novo.query('form_reaj=="FR-003"')
fr003["preco_previsto"] = ((fr003["Valor Unitario"] * 
fr003["Estanho F"])/100 + 
fr003["Valor Unitario"]).round(2)

fr004 = df_preco_novo.query('form_reaj=="FR-004"')
fr004["preco_previsto"] = ((fr004["Valor Unitario"] * 
fr004["IGPM F"])/100 + 
fr004["Valor Unitario"]).round(2)

fr005 = df_preco_novo.query('form_reaj=="FR-005"')
fr005["preco_previsto"] = ((fr005["Valor Unitario"] * 
fr005["Petro F"])/100 + 
fr005["Valor Unitario"]).round(2)

fr006 = df_preco_novo.query('form_reaj=="FR-006"')
fr006["preco_previsto"] = (
                           ((fr006["Valor Unitario"] * fr006["Dolar F"])/100*0.5)+
                           ((fr006["Valor Unitario"] * fr006["IGPM F"])/100*0.5)+
                             fr006["Valor Unitario"]
                        ).round(2)

fr007 = df_preco_novo.query('form_reaj=="FR-007"')
fr007["preco_previsto"] = (
                           ((fr007["Valor Unitario"] * fr007["Petro F"])/100*0.5)+
                           ((fr007["Valor Unitario"] * fr007["Soja F"])/100*0.5)+
                             fr007["Valor Unitario"]
                        ).round(2)

df_preco_previsto = pd.concat([fr001,fr002,fr003,fr004,fr005,fr006,fr007])
df_preco_previsto = df_preco_previsto[['Material', 'Pedido', 'Valor Unitario', 'preco_previsto']]
df_preco_previsto = df_preco_previsto.sort_values(by='Material')
df_preco_previsto.tail(15)
```

# <font color=orange>4) Unificando todas as informações em um Dataset</font>
***

```python
# Calculando a diferença entre "Preço pago última compra" e o "Preço Previsto" 
## Objetivo é verificar se a previsão é de aumento ou queda no preço.

diferenca = df_preco_previsto[{'Material', 'Pedido', 'Valor Unitario', 'preco_previsto'}]
diferenca= diferenca.rename(columns={'Valor Unitario': 'ultimo_preco'})
diferenca = diferenca[['Material', 'Pedido', 'ultimo_preco', 'preco_previsto']]
diferenca["diferenca"] = ((diferenca["preco_previsto"] - 
                         diferenca["ultimo_preco"]
                        )).round(2)
diferenca
```

```python
# Criando dataset com as informações, por RC

df_rcs = requisicoes[{'Req. Compra', 'Material', 'Qtde', 'UM', 'Data doc.'}]

df_rcs = pd.merge(df_rcs, diferenca, on=['Material'], how='left')
df_rcs = df_rcs.drop(columns=['Data doc.','Pedido'])
df_rcs["montante"] = ((df_rcs["diferenca"] * 
                         df_rcs["Qtde"]
                        )).round(2)
df_rcs = df_rcs[['Req. Compra', 'Material', 'Qtde', 'UM', 'ultimo_preco', 'preco_previsto', 'diferenca', 'montante']]
df_rcs
```

# <font color=orange>5) Verificando quais RC's possuem contratos com possibilidade de aumento de preço</font>
***

```python
df_contratos = contratos[{'Contrato', 'Material', 'Data Validade', 'Saldo', 'Fornecedor'}]
ultimo_contrato = df_contratos.groupby(["Material"])['Contrato'].max().reset_index()
df_contratos = pd.merge(df_contratos, ultimo_contrato, on=['Material'], how='left')

def f(row):
    if row['Contrato_x'] == row['Contrato_y']:
        val = 0
    elif row['Contrato_x'] > row['Contrato_y']:
        val = 1
    else:
        val = -1
    return val  

df_contratos["retirar"] = df_contratos.apply(f, axis=1)
df_contratos= df_contratos.rename(columns={'Contrato_x': 'Contrato'})
df_contratos = df_contratos.query('retirar==0')
df_contratos = df_contratos.drop(columns=['Contrato_y', 'retirar'])
df_contratos
```

```python
# Verificando RC's que possuam contratos, saldo e período de vigência corretos

rcs_1 = df_contratos
rcs_1 = pd.merge(df_rcs, rcs_1, on=['Material'], how='right')
rcs_1["total_previsto"] = ((rcs_1["Qtde"] * 
                         rcs_1["ultimo_preco"]
                        )).round(2)
rcs_1["saldo_novo"] = ((rcs_1["Saldo"] - 
                         rcs_1["total_previsto"]
                        )).round(2)
rcs_1 = rcs_1.query('montante>=0')
rcs_1 = rcs_1.query('saldo_novo>=0')
rcs_1
```

```python
# Verificando RC's que possuam contratos, período de vigência corretos, mas que não tem saldo.

rcs_2 = df_contratos
rcs_2 = pd.merge(df_rcs, rcs_2, on=['Material'], how='right')
rcs_2["total_previsto"] = ((rcs_2["Qtde"] * 
                         rcs_2["ultimo_preco"]
                        )).round(2)
rcs_2["saldo_novo"] = ((rcs_2["Saldo"] - 
                         rcs_2["total_previsto"]
                        )).round(2)
rcs_2 = rcs_2.query('montante>=0')
rcs_2 = rcs_2.query('saldo_novo<0')
rcs_2
```

# <font color=orange>6) RC's que possuem contratos, mas que poderiam ter redução de preços</font>
***

```python
rcs_3 = df_contratos
rcs_3 = pd.merge(df_rcs, rcs_3, on=['Material'], how='right')
rcs_3["total_previsto"] = ((rcs_3["Qtde"] * 
                         rcs_3["ultimo_preco"]
                        )).round(2)
rcs_3["saldo_novo"] = ((rcs_3["Saldo"] - 
                         rcs_3["total_previsto"]
                        )).round(2)
rcs_3 = rcs_3.query('montante<0')
rcs_3
```

# <font color=orange>7) RC's que não possuem contratos, mas que estão prontas para emissão de pedido</font>
***

```python
# RC's que estão sem contrato, mas que tem previsão de alta nos preços. Portanto melhor indicação de comprar conforme último pedido.

rcs_4 = contratos[{'Contrato', 'Material', 'Data Validade', 'Saldo'}]
rcs_4 = pd.merge(df_rcs, rcs_4, on=['Material'], how='left')
rcs_4 = rcs_4.query('montante>=0')
rcs_4.fillna("Fica", inplace = True)
rcs_4 = rcs_4.query('Saldo=="Fica"')
rcs_4 = rcs_4[['Req. Compra', 'Material', 'Qtde', 'UM', 'ultimo_preco', 'preco_previsto', 'diferenca', 'montante']]
rcs_4 = rcs_4.sort_values(by='Req. Compra')
rcs_4
```

# <font color=orange>8)  RC's que não possuem contratos, que devem ser designadas para cotação</font>
***

```python
# RC's que estão sem contrato e que tem previsão de queda de preços. Designadas para abertura de cotação

rcs_5 = contratos[{'Contrato', 'Material', 'Data Validade', 'Saldo'}]
rcs_5 = pd.merge(df_rcs, rcs_5, on=['Material'], how='left')
rcs_5 = rcs_5.query('montante<0')
rcs_5.fillna("Fica", inplace = True)
rcs_5 = rcs_5.query('Saldo=="Fica"')
rcs_5 = rcs_5[['Req. Compra', 'Material', 'Qtde', 'UM', 'ultimo_preco', 'preco_previsto', 'diferenca', 'montante']]
rcs_5 = rcs_5.sort_values(by='Req. Compra')
rcs_5
```

# <font color=orange>13) RC's que devem ser enviadas para cotação</font>
***

```python
rcs_5 = rcs_5[{'Req. Compra', 'Material', 'Qtde', 'UM'}]
rcs_6 = rcs_6[{'Req. Compra', 'Material', 'Qtde', 'UM'}]
cotacao = pd.concat([rcs_5,rcs_6])
cotacao.head(3)
```

```python
df_condicoes = condicoes[{'Material', 'Grp Merc'}]
agrupando_rcs = pd.merge(cotacao, df_condicoes, on=['Material'], how='left')
agrupando_rcs.head(10)
```

```python
groupby1 = agrupando_rcs.groupby(["Grp Merc", "Material"])['Req. Compra'].count().reset_index()
groupby2 = agrupando_rcs.groupby(["Grp Merc", "Req. Compra"])['Material'].count().reset_index()
groupby3 = agrupando_rcs.groupby(["Grp Merc"])['Req. Compra'].count().reset_index()
agrupando = pd.merge(groupby1, groupby3, on=['Grp Merc'], how='left')
agrupando = pd.merge(agrupando, groupby2, on=['Grp Merc'], how='left')

def r(row):
    if row['Req. Compra_x'] == row['Req. Compra_y'] == row['Material_y']:
        val = 0
    else:
        val = 1
    return val  

agrupando["conc"] = agrupando.apply(r, axis=1)
agrupando= agrupando.rename(columns={'Material_x': 'Material'})
agrupando= agrupando.rename(columns={'Req. Compra_x': 'Req. Compra_Z'})

agrupando.head(3)
```

```python
#Selecionando apenas as RC's que não precisam ser agrupadas para cotação:

enviar_cotacao = agrupando

enviar_cotacao = enviar_cotacao.query('conc==0')
enviar_cotacao = enviar_cotacao.rename(columns={'Material_x': 'Material'})
enviar_cotacao = pd.merge(agrupando_rcs, enviar_cotacao, on=['Grp Merc'], how='left')
enviar_cotacao = enviar_cotacao.dropna()
enviar_cotacao = enviar_cotacao.rename(columns={'Req. Compra': 'OUT'})
enviar_cotacao = enviar_cotacao.rename(columns={'Req. Compra_x': 'Req. Compra'})
enviar_cotacao = enviar_cotacao.rename(columns={'Material_x': 'Material'})
rcs_7 = enviar_cotacao[{'Material', 'Req. Compra', 'Qtde', 'UM'}]

rcs_7
```

```python
#Consolidando as cotações:

consolidar = agrupando

consolidar = consolidar.query('conc==1')
consolidar = consolidar.rename(columns={'Material_x': 'Material'})
consolidar = consolidar[{'Grp Merc', 'Req. Compra'}]
consolidar = consolidar.drop_duplicates()

rcs_8 = consolidar
rcs_8["Consolidar"] = 1
rcs_8 = rcs_8[{'Grp Merc', 'Consolidar'}]
rcs_8 = rcs_8.drop_duplicates()
rcs_8 = pd.merge(agrupando_rcs, rcs_8, on=['Grp Merc'], how='left')
rcs_8 = rcs_8.dropna()
rcs_8 = rcs_8.sort_values('Grp Merc')
rcs_8 = rcs_8[{'Req. Compra', 'Material', 'Qtde', 'UM', 'Grp Merc'}]
rcs_8
```
